```python
import numpy as np
import pandas as pd
import ast
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
jobs = pd.read_csv('internship companies.csv')
```

```python
jobs
```

```python
# company
# internship
# location
# start date
# duration
```

```python
jobs.isnull().sum()
```

```python
jobs.dropna(inplace=True)
```

```python
jobs.duplicated().sum()
```

```python
jobs['company'] = jobs['company'].apply(lambda x:x.replace(" ", ""))
jobs['internship'] = jobs['internship'].apply(lambda x:x.replace(" ", ""))
jobs['location'] = jobs['location'].apply(lambda x:x.replace(" ", ""))
jobs['start date'] = jobs['start date'].apply(lambda x:x.replace(" ", ""))
jobs['duration'] = jobs['duration'].apply(lambda x:x.replace(" ", ""))
```

```python
jobs
```

```python
jobs['tags'] = jobs['company'] + " " + jobs['internship'] + " " + jobs['location'] + " " + jobs['start date'] + " " + jobs['duration']
```

```python
jobs
```

```python
jobs['tags'] = jobs['tags'].apply(lambda x:x.lower())
```

```python
jobs
```

```python
cv = CountVectorizer(max_features=100, stop_words='english')
```

```python
vectors = cv.fit_transform(jobs['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
vectors
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:11]
```

```python
def recommend(job):
    job_index = jobs[jobs['internship'] == job].index[0]
    distances = similarity[job_index]
    jobs_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:100]
    
    for i in jobs_list:
        print(jobs.iloc[i[0]].internship)
```

```python
recommend('WebDevelopment')
```
